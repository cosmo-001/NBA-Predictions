# regular players season

```python
import pandas as pd
from sklearn.svm import OneClassSVM
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('/content/player_regular_season_career.txt')

# Extract the "pts," "reb," "asts," and "player_name" columns for outlier detection
features = data[['pts', 'reb', 'asts']]
player_names = data[['firstname', 'lastname']]

# Initialize the One-Class SVM model for outlier detection
svm = OneClassSVM(nu=0.05, kernel="rbf")

# Fit the model to the data
svm.fit(features)

# Predict inliers and outliers
outliers = svm.predict(features)

# Extract outstanding players
outstanding_players = data[outliers == -1]

outstanding_players = outstanding_players.drop_duplicates(subset=['firstname', 'lastname'])

print(f"Outstanding Players:")
for index, row in outstanding_players.iterrows():
    player_index = index
    player_name = row['firstname']
    player_surname = row['lastname']
    print(f" {player_index} {player_name} {player_surname}")

# Count  of outstanding players
count_outstanding_players = len(outstanding_players)
print(f"Count of Outstanding Players: {count_outstanding_players}")
```

```python
import pandas as pd
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Read Data from Text File
data = pd.read_csv('/content/player_allstar.txt')

#Data Preprocessing
selected_features = data[['pts', 'reb', 'asts']].copy()
player_names = data[['firstname', 'lastname']].copy()

# Remove rows with missing values in any of the selected features
selected_features.dropna(inplace=True)
player_names = player_names.loc[selected_features.index]

# Scale/normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(selected_features)

#Train the One-Class SVM Model with adjusted parameters
model = OneClassSVM(nu=0.01, kernel='rbf')
model.fit(X_scaled)

# Predict Outliers
predictions = model.predict(X_scaled)
outliers = player_names[predictions == -1]


outliers = outliers.drop_duplicates()

print("Outstanding Players:")
print(outliers)

# Count of outstanding players
count_outstanding_players = len(outliers)
print(f"Count of Outstanding Players: {count_outstanding_players}")



```

# Player playoffs


```python
import pandas as pd
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

#Read Data from Text File
data = pd.read_csv('/content/player_playoffs_career.txt')  # Replace 'your_data.txt' with the actual file path

# Data Preprocessing
selected_features = data[['pts', 'reb', 'asts']].copy()

# Remove rows with missing values in any of the selected features
selected_features.dropna(inplace=True)

# Extract player names
player_names = data[['firstname', 'lastname']].loc[selected_features.index]

# Scale/Normalize the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(selected_features)

#Train the One-Class SVM Model with adjusted parameters
model = OneClassSVM(nu=0.01, kernel='rbf')
model.fit(X_scaled)

# Predict Outliers
predictions = model.predict(X_scaled)

# Extract the names of players identified as outliers
outlier_names = player_names[predictions == -1]

# Remove duplicate player names from the list of outlier_names
outlier_names = outlier_names.drop_duplicates()

# Print the names of the outlier players
print("Outstanding Players:")
print(outlier_names)

# Count the number of outliers
num_outliers = len(outlier_names)
print(f"Count of Outstanding Players: {num_outliers}")

```

# Team win/loss prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix

# Load data
data = pd.read_csv('team_season.txt')

# Select features and target
X = data[['o_fgm', 'o_fga', 'o_ftm', 'o_fta', 'o_oreb', 'o_dreb', 'o_reb', 'o_asts', 'o_pf', 'o_stl', 'o_to',
          'o_blk', 'o_3pm', 'o_3pa', 'pace', 'd_fgm', 'd_fga', 'd_ftm', 'd_fta', 'd_oreb', 'd_dreb', 'd_reb',
          'd_asts', 'd_pf', 'd_stl', 'd_to', 'd_blk', 'd_3pm', 'd_3pa', 'pace']]
y = data['won']

# Create a binary target variable: 1 for wins, 0 for losses
y_binary = (y > y.median()).astype(int)

# Standardize the numeric features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Create and train an SVM model
svm_classifier = SVC(random_state=42)
svm_classifier.fit(X_train, y_train)

# Input team names from the user
team_a_name = input("Enter the name of the first team: ")
team_b_name = input("Enter the name of the second team: ")

# Filter data for the selected teams
team_a_data = data[data['team'] == team_a_name][['o_fgm', 'o_fga', 'o_ftm', 'o_fta', 'o_oreb', 'o_dreb', 'o_reb',
                                                 'o_asts', 'o_pf', 'o_stl', 'o_to', 'o_blk', 'o_3pm', 'o_3pa',
                                                 'pace', 'd_fgm', 'd_fga', 'd_ftm', 'd_fta', 'd_oreb', 'd_dreb',
                                                 'd_reb', 'd_asts', 'd_pf', 'd_stl', 'd_to', 'd_blk', 'd_3pm',
                                                 'd_3pa', 'pace']]
team_b_data = data[data['team'] == team_b_name][['o_fgm', 'o_fga', 'o_ftm', 'o_fta', 'o_oreb', 'o_dreb', 'o_reb',
                                                 'o_asts', 'o_pf', 'o_stl', 'o_to', 'o_blk', 'o_3pm', 'o_3pa',
                                                 'pace', 'd_fgm', 'd_fga', 'd_ftm', 'd_fta', 'd_oreb', 'd_dreb',
                                                 'd_reb', 'd_asts', 'd_pf', 'd_stl', 'd_to', 'd_blk', 'd_3pm',
                                                 'd_3pa', 'pace']]

# Check if the DataFrames are empty
if team_a_data.empty or team_b_data.empty:
    print("One or both teams not found in the dataset.")
else:
    # Standardize the team data
    team_a_data_standardized = scaler.transform(team_a_data)
    team_b_data_standardized = scaler.transform(team_b_data)

    # Predict the outcome of the matchup
    outcome = svm_classifier.predict(team_a_data_standardized - team_b_data_standardized)

    if outcome[0] == 1:
        print(f"{team_a_name} wins")
        print(f"{team_b_name} loses")
    else:
        print(f"{team_b_name} wins")
        print(f"{team_a_name} loses")

    # Evaluate the model on the test set
    test_predictions = svm_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, test_predictions)
    print(f"Accuracy on Test Set: {accuracy}")

    # Additional metrics
    conf_matrix = confusion_matrix(y_test, test_predictions)
    print("Confusion Matrix:")
    print(conf_matrix)
```
