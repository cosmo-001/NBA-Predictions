```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

player_playoff='/Users/Pashlene/Desktop/Modified_player_playoffs.csv'
player_playoff_career='/Users/Pashlene/Desktop/Modified_player_playoffs_career.csv'

player_playoff_df=pd.read_csv(player_playoff)
player_playoff_career_df=pd.read_csv(player_playoff_career)
```

```python
player_playoff_df.fillna(0, inplace=True)
player_playoff_df.head()
```

```python
performance_metrics = ['pts', 'oreb', 'dreb', 'asts', 'stl', 'blk']

# We will handle missing values by filling them with the median of the column
df_filled = player_playoff_df[performance_metrics].fillna(0)

iforest = IsolationForest(n_estimators = 100, contamination = 0.03, max_samples ='auto')
prediction = iforest.fit_predict(df_filled)
print(prediction)
print("Number of outliers detected: {}".format(prediction[prediction < 0].sum()))
print("Number of normal samples detected: {}".format(prediction[prediction > 0].sum()))
player_playoff_df['anomaly_score']=prediction
```

```python
outstanding_players = player_playoff_df[player_playoff_df['anomaly_score'] == -1]
outstanding_players
```
