```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

reg_player_season = '/Users/Pashlene/Desktop/Modified_player_regular_season.csv'
reg_player_season_career = '/Users/Pashlene/Desktop/Modified_player_regular_season_career.csv'

# Read the CSV file into a DataFrame
reg_player_season_df = pd.read_csv(reg_player_season)
reg_player_season_career_df= pd.read_csv(reg_player_season_career)
```

```python
reg_player_season_df.head()
```

```python
performance_metrics = ['pts', 'oreb', 'dreb', 'asts', 'stl', 'blk']

# We will handle missing values by filling them with the median of the column
df_filled = reg_player_season_df[performance_metrics].fillna(0)

iforest = IsolationForest(n_estimators = 100, contamination = 0.03, max_samples ='auto')
prediction = iforest.fit_predict(df_filled)
print(prediction)
print("Number of outliers detected: {}".format(prediction[prediction < 0].sum()))
print("Number of normal samples detected: {}".format(prediction[prediction > 0].sum()))
reg_player_season_df['anomaly_score']=prediction
```

```python
outstanding_players = reg_player_season_df[reg_player_season_df['anomaly_score'] == -1]
outstanding_players
```
